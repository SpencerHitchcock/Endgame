```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from html.parser import HTMLParser
```

```python
#get data
data = requests.get('https://www.the-numbers.com/movie/Iron-Man#tab=box-office')
#load into bs4
soup = BeautifulSoup(data.text, 'html.parser')

data = []
div = soup.find('div' , {'id': 'box_office_chart'})
table = div.find('table')
tbody = table.find('tbody')

print(table)

```

```python

for table_row in soup.select("table"):
    cells = table_row.find_all(['td'])
    date = cells[0].text.strip()
    rank = cells[1].text.strip()
    gross = cells[5].text.strip()  
    week = cells[6].text.strip()

    
    movie = pd.DataFrame(["Date": date,
              "Rank": rank,
              "Gross": gross,
              "Week": week])
    data.append(movie)
    print (date, rank, gross, week)
```
